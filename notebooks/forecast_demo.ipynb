# Tekimax Forecast Demo Notebook

This notebook demonstrates the forecasting and what-if APIs used in the platform demo.

```python
import requests

base_url = "http://localhost:8001"
payload = {
    "series": [1200, 1350, 1425, 1500, 1580, 1660],
    "periods": 3,
    "model": "moving_average"
}
requests.post(f"{base_url}/forecast", json=payload, timeout=30).json()
```

```python
scenario = {
    "series": [1200, 1350, 1425, 1500, 1580, 1660],
    "periods": 3,
    "change_percent": -10
}
requests.post(f"{base_url}/what-if", json=scenario, timeout=30).json()
```
